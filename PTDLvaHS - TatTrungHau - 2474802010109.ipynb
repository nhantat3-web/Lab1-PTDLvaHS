```python
# Câu 3: Tải dữ liệu và in ra 10 dòng đầu tiên, 10 dòng cuối cùng

import pandas as pd

# Đọc file dữ liệu
df = pd.read_csv("dulieuxettuyendaihoc.csv")

# In 10 dòng đầu tiên
print("10 dòng đầu tiên:")
display(df.head(10))

# In 10 dòng cuối cùng
print("10 dòng cuối cùng:")
display(df.tail(10))
```

##  Câu 4: Thống kê dữ liệu thiếu cho cột dân tộc DT

```python
# Mặc định thiếu thì điền giá trị 0

# Lập bảng tần số và tần suất trước khi xử lý
tan_so_DT = df["DT"].value_counts(dropna=False)
tan_suat_DT = df["DT"].value_counts(dropna=False, normalize=True) * 100

bang_thong_ke_DT = pd.DataFrame({
    "Tần số": tan_so_DT,
    "Tần suất (%)": tan_suat_DT
})

print("BẢNG TẦN SỐ, TẦN SUẤT CỦA CỘT DT TRƯỚC KHI XỬ LÝ")
display(bang_thong_ke_DT)

# Thống kê dữ liệu thiếu
so_luong_thieu_DT = df["DT"].isnull().sum()
ti_le_thieu_DT = df["DT"].isnull().mean() * 100

print("Số lượng dữ liệu thiếu của cột DT:", so_luong_thieu_DT)
print("Tỉ lệ dữ liệu thiếu của cột DT (%):", ti_le_thieu_DT)

# Khảo sát dữ liệu riêng biệt bằng unique
du_lieu_rieng_biet_DT = df["DT"].unique()
so_du_lieu_rieng_biet_DT = df["DT"].nunique(dropna=True)

print("Các dữ liệu riêng biệt của cột DT:")
print(du_lieu_rieng_biet_DT)

print("Số lượng dữ liệu riêng biệt của cột DT:", so_du_lieu_rieng_biet_DT)

# Thay thế dữ liệu thiếu bằng giá trị 0
df["DT"] = df["DT"].fillna(0)

print("\nSAU KHI ĐIỀN DỮ LIỆU THIẾU BẰNG 0")
print("Số lượng dữ liệu thiếu của cột DT:", df["DT"].isnull().sum())

# Lập lại bảng tần số và tần suất sau khi xử lý
tan_so_DT_sau = df["DT"].value_counts(dropna=False)
tan_suat_DT_sau = df["DT"].value_counts(dropna=False, normalize=True) * 100

bang_thong_ke_DT_sau = pd.DataFrame({
    "Tần số": tan_so_DT_sau,
    "Tần suất (%)": tan_suat_DT_sau
})

print("BẢNG TẦN SỐ, TẦN SUẤT CỦA CỘT DT SAU KHI XỬ LÝ")
display(bang_thong_ke_DT_sau)
```

## 5. Thống kê dữ liệu thiếu cho biến T1 và hiệu chỉnh dữ liệu, lưu ý việc thay thế dữ liệu thiếu sử dụng phương pháp Mean.

```python
# Hiệu chỉnh dữ liệu thiếu bằng phương pháp điền giá trị Mean

# 1. Lập bảng tần số và tần suất trước khi xử lý
tan_so_T1 = df["T1"].value_counts(dropna=False)
tan_suat_T1 = df["T1"].value_counts(dropna=False, normalize=True) * 100

bang_thong_ke_T1 = pd.DataFrame({
    "Tần số": tan_so_T1,
    "Tần suất (%)": tan_suat_T1
})

print("BẢNG TẦN SỐ, TẦN SUẤT CỦA CỘT T1 TRƯỚC KHI XỬ LÝ")
display(bang_thong_ke_T1)

# 2. Thống kê dữ liệu thiếu
so_luong_thieu_T1 = df["T1"].isnull().sum()
ti_le_thieu_T1 = df["T1"].isnull().mean() * 100

print("Số lượng dữ liệu thiếu của cột T1:", so_luong_thieu_T1)
print("Tỉ lệ dữ liệu thiếu của cột T1 (%):", ti_le_thieu_T1)

# 3. Khảo sát dữ liệu riêng biệt bằng unique
du_lieu_rieng_biet_T1 = df["T1"].unique()
so_du_lieu_rieng_biet_T1 = df["T1"].nunique(dropna=True)

print("Các dữ liệu riêng biệt của cột T1:")
print(du_lieu_rieng_biet_T1)

print("Số lượng dữ liệu riêng biệt của cột T1:", so_du_lieu_rieng_biet_T1)

# 4. Tính giá trị Mean của T1
mean_T1 = df["T1"].mean()

print("Giá trị Mean của cột T1:", mean_T1)

# 5. Thay thế dữ liệu thiếu bằng giá trị Mean
df["T1"] = df["T1"].fillna(mean_T1)

print("\nSAU KHI ĐIỀN DỮ LIỆU THIẾU BẰNG MEAN")
print("Số lượng dữ liệu thiếu của cột T1:", df["T1"].isnull().sum())

# 6. Lập lại bảng tần số và tần suất sau khi xử lý
tan_so_T1_sau = df["T1"].value_counts(dropna=False)
tan_suat_T1_sau = df["T1"].value_counts(dropna=False, normalize=True) * 100

bang_thong_ke_T1_sau = pd.DataFrame({
    "Tần số": tan_so_T1_sau,
    "Tần suất (%)": tan_suat_T1_sau
})

print("BẢNG TẦN SỐ, TẦN SUẤT CỦA CỘT T1 SAU KHI XỬ LÝ")
display(bang_thong_ke_T1_sau)

# 7. Kết luận
print("\nKết luận:")
print("Đã thống kê dữ liệu thiếu của cột T1.")
print("Dữ liệu thiếu của cột T1 đã được thay thế bằng giá trị Mean.")
```

## 6. Hãy thực hiện xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại. 

```python
# Phương pháp: điền dữ liệu thiếu bằng giá trị Mean của từng cột

# Danh sách tất cả các cột điểm số
cot_diem = [
    "T1", "L1", "H1", "S1", "V1", "X1", "D1", "N1",
    "T2", "L2", "H2", "S2", "V2", "X2", "D2", "N2",
    "T3", "L3", "H3", "S3", "V3", "X3", "D3", "N3",
    "T4", "L4", "H4", "S4", "V4", "X4", "D4", "N4",
    "T5", "L5", "H5", "S5", "V5", "X5", "D5", "N5",
    "T6", "L6", "H6", "S6", "V6", "X6", "D6", "N6",
    "DH1", "DH2", "DH3"
]

# Lọc lại các cột có trong DataFrame để tránh lỗi
cot_diem = [cot for cot in cot_diem if cot in df.columns]

# Thống kê dữ liệu thiếu trước khi xử lý
bang_thieu_truoc = pd.DataFrame({
    "Cột": cot_diem,
    "Số lượng thiếu": [df[cot].isnull().sum() for cot in cot_diem],
    "Tỉ lệ thiếu (%)": [df[cot].isnull().mean() * 100 for cot in cot_diem],
    "Số giá trị riêng biệt": [df[cot].nunique(dropna=True) for cot in cot_diem]
})

print("BẢNG THỐNG KÊ DỮ LIỆU THIẾU TRƯỚC KHI XỬ LÝ")
display(bang_thieu_truoc)

# Thay thế dữ liệu thiếu bằng Mean cho từng cột điểm
for cot in cot_diem:
    mean_cot = df[cot].mean()
    df[cot] = df[cot].fillna(mean_cot)

# Thống kê dữ liệu thiếu sau khi xử lý
bang_thieu_sau = pd.DataFrame({
    "Cột": cot_diem,
    "Số lượng thiếu": [df[cot].isnull().sum() for cot in cot_diem],
    "Tỉ lệ thiếu (%)": [df[cot].isnull().mean() * 100 for cot in cot_diem],
    "Số giá trị riêng biệt": [df[cot].nunique(dropna=True) for cot in cot_diem]
})

print("BẢNG THỐNG KÊ DỮ LIỆU THIẾU SAU KHI XỬ LÝ")
display(bang_thieu_sau)

# Kiểm tra tổng số dữ liệu thiếu trong các cột điểm
print("Tổng số dữ liệu thiếu còn lại trong các cột điểm:", df[cot_diem].isnull().sum().sum())

print("\nKết luận:")
print("Đã xử lý dữ liệu thiếu cho tất cả các biến điểm số bằng phương pháp điền giá trị Mean.")
```

## Câu 7: Tạo biến TBM1, TBM2, TBM3

```python
# TBM1: điểm trung bình lớp 10
# TBM2: điểm trung bình lớp 11
# TBM3: điểm trung bình lớp 12

def tinh_tbm_hoc_ky(ky):
    return (
        df[f"T{ky}"] * 2 +
        df[f"L{ky}"] +
        df[f"H{ky}"] +
        df[f"S{ky}"] +
        df[f"V{ky}"] * 2 +
        df[f"X{ky}"] +
        df[f"D{ky}"] +
        df[f"N{ky}"]
    ) / 10

# Tính điểm trung bình từng học kỳ
tbm_hk1 = tinh_tbm_hoc_ky(1)
tbm_hk2 = tinh_tbm_hoc_ky(2)
tbm_hk3 = tinh_tbm_hoc_ky(3)
tbm_hk4 = tinh_tbm_hoc_ky(4)
tbm_hk5 = tinh_tbm_hoc_ky(5)
tbm_hk6 = tinh_tbm_hoc_ky(6)

# Tính điểm trung bình từng năm
# HK2 nhân hệ số 2
df["TBM1"] = ((tbm_hk1 + tbm_hk2 * 2) / 3).round(2)
df["TBM2"] = ((tbm_hk3 + tbm_hk4 * 2) / 3).round(2)
df["TBM3"] = ((tbm_hk5 + tbm_hk6 * 2) / 3).round(2)

print("KẾT QUẢ TẠO BIẾN TBM1, TBM2, TBM3")
display(df[["TBM1", "TBM2", "TBM3"]].head(10))

print("Thống kê mô tả của TBM1, TBM2, TBM3:")
display(df[["TBM1", "TBM2", "TBM3"]].describe())

print("Kết luận: Đã tạo thành công 3 biến TBM1, TBM2, TBM3 cho lớp 10, 11, 12.")
```

## 8. Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp 10, 11, 12

```python
# Hàm xếp loại học lực
def xep_loai(tbm):
    if tbm < 5.0:
        return "Y"      # Yếu
    elif tbm < 6.5:
        return "TB"     # Trung bình
    elif tbm < 8.0:
        return "K"      # Khá
    elif tbm < 9.0:
        return "G"      # Giỏi
    else:
        return "XS"     # Xuất sắc

# Tạo các biến xếp loại
df["XL1"] = df["TBM1"].apply(xep_loai)
df["XL2"] = df["TBM2"].apply(xep_loai)
df["XL3"] = df["TBM3"].apply(xep_loai)

# Hiển thị kết quả 10 dòng đầu
print("KẾT QUẢ TẠO BIẾN XL1, XL2, XL3")
display(df[["TBM1", "XL1", "TBM2", "XL2", "TBM3", "XL3"]].head(10))

# Thống kê tần số xếp loại
print("Tần số xếp loại XL1:")
display(df["XL1"].value_counts())

print("Tần số xếp loại XL2:")
display(df["XL2"].value_counts())

print("Tần số xếp loại XL3:")
display(df["XL3"].value_counts())

print("Kết luận: Đã tạo thành công các biến xếp loại XL1, XL2, XL3.")
```

## 9. Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp 10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương pháp Min-Max Normalization 

```python
# Chuyển điểm trung bình từ thang điểm 10 sang thang điểm 4
# Công thức Min-Max Normalization:
# US_TBM = (TBM / 10) * 4

# Tạo các biến US_TBM
df["US_TBM1"] = (df["TBM1"] / 10) * 4
df["US_TBM2"] = (df["TBM2"] / 10) * 4
df["US_TBM3"] = (df["TBM3"] / 10) * 4

# Làm tròn 2 chữ số thập phân
df["US_TBM1"] = df["US_TBM1"].round(2)
df["US_TBM2"] = df["US_TBM2"].round(2)
df["US_TBM3"] = df["US_TBM3"].round(2)

# Hiển thị kết quả 10 dòng đầu
print("KẾT QUẢ TẠO BIẾN US_TBM1, US_TBM2, US_TBM3")
display(df[[
    "TBM1", "US_TBM1",
    "TBM2", "US_TBM2",
    "TBM3", "US_TBM3"
]].head(10))

# Thống kê mô tả các biến mới
print("THỐNG KÊ MÔ TẢ CÁC BIẾN US_TBM")
display(df[["US_TBM1", "US_TBM2", "US_TBM3"]].describe())

print("Kết luận: Đã tạo thành công các biến US_TBM1, US_TBM2, US_TBM3 bằng phương pháp Min-Max Normalization.")
```

## Câu 10: Tạo biến KQXT - Kết quả xét tuyển

```python
# Quy tắc:
# Nếu khối A hoặc A1: điểm xét tuyển = (DH1*2 + DH2 + DH3) / 4
# Nếu khối B: điểm xét tuyển = (DH1 + DH2*2 + DH3) / 4
# Các khối còn lại: điểm xét tuyển = (DH1 + DH2 + DH3) / 3
# Nếu điểm xét tuyển >= 5 thì KQXT = 1, ngược lại KQXT = 0

def tinh_kqxt(row):
    khoi = str(row["KT"]).strip()
    dh1 = row["DH1"]
    dh2 = row["DH2"]
    dh3 = row["DH3"]
    
    if khoi in ["A", "A1"]:
        diem_xet_tuyen = (dh1 * 2 + dh2 + dh3) / 4
    elif khoi == "B":
        diem_xet_tuyen = (dh1 + dh2 * 2 + dh3) / 4
    else:
        diem_xet_tuyen = (dh1 + dh2 + dh3) / 3
    
    if diem_xet_tuyen >= 5:
        return 1
    else:
        return 0

# Tạo biến KQXT
df["KQXT"] = df.apply(tinh_kqxt, axis=1)

# Hiển thị kết quả 10 dòng đầu
print("KẾT QUẢ TẠO BIẾN KQXT")
display(df[["KT", "DH1", "DH2", "DH3", "KQXT"]].head(10))

# Thống kê tần số kết quả xét tuyển
print("Tần số kết quả xét tuyển:")
display(df["KQXT"].value_counts())

print("Kết luận: Đã tạo thành công biến KQXT.")
```
